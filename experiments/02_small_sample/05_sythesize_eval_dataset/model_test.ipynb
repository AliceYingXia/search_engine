```python
import requests
import json

# Define the endpoint URL
url = 'https://ai-gateway-int.awstf.workato.com/v1/embeddings'

# Set up the headers
headers = {
    'Authorization': 'Bearer sk-chat-api-1moNkcolQx8Sn2EKcSsKRjtxvv3m4lzS',
    'Content-Type': 'application/json'
}

# Define the payload (the data you are sending)
data = {
    "model": "text-embedding-3-large",
    "input": "Hello from ai-gateway"
}

try:
    # Make the POST request
    # Use 'json=' instead of 'data=' to automatically set the content-type 
    # and serialize the dictionary to a JSON string
    response = requests.post(url, headers=headers, json=data)

    # Check if the request was successful
    response.raise_for_status()

    # Parse and print the JSON response
    result = response.json()
    print(json.dumps(result, indent=2))

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
except Exception as err:
    print(f"An error occurred: {err}")
```
